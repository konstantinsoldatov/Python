## 1. Почему Python часто выбирают в физике конденсированного состояния и других смежных областях.

Python удобен тем, что позволяет быстро проверить идею вычислительным экспериментом без долгой подготовки проекта и компиляции. Это особенно полезно в физике, где часто нужно быстро построить модель, промоделировать зависимость или обработать массив измерений. У языка есть большая экосистема библиотек для численных методов, статистики, визуализации и символьной математики, и почти всё это хорошо интегрируется с Jupyter. Наконец, ноутбуки поддерживают воспроизводимость: вы можете показать и текст рассуждений, и формулы, и код, и результат вычислений в одном документе.

**Очень важно**: Python — интерпретируемый язык с **динамической типизацией**.
 Это означает, что тип данных не закреплён за именем переменной навсегда, а привязан к объекту - значению на которое переменная ссылается. 
Когда вы пишете x = 3, имя x начинает ссылаться на объект типа int. Если потом сделать x = 3.0, то то же имя x начнёт ссылаться уже на объект типа float. Никакого «переприсваивания типа переменной» как отдельной операции нет: меняется объект, на который указывает имя.
В Python переменная — это не «ячейка памяти фиксированного типа», как в языках C и C++, а ссылка/ярлык на объект в куче. Поэтому выражение type(x) фактически отвечает на вопрос «какого типа объект, на который сейчас ссылается имя x?». Это тесно связано с моделью памяти: присваивание = связывает имя с объектом, а не копирует «значение в переменную».

Динамическая типизация отличается от статической (например, в C/C++/Java), где типы многих сущностей фиксируются заранее (на этапе компиляции) и нельзя просто так присвоить переменной другого типа без явного преобразования или другой переменной. В статически типизированном языке конструкция вроде «вчера x был int, а сегодня x стал string» обычно невозможна (или требует обобщённых типов/вариантов). В Python это нормально, потому что тип всегда у объекта, а имя может указывать на любой объект.

Код становится гибким и быстро пишется, но повышается риск «тихих» ошибок из-за неожиданных типов данных на входе. Например, если функция ожидает число, а вы случайно передали строку из файла, ошибка проявится не при запуске программы как таковой, а когда интерпретатор дойдет до места с ошибкой.


````python
x = 3
print(x,": тип данных", type(x))   

x = 3.0
print(x,": тип данных",  type(x))  

x = "3"
print(x,": тип данных",  type(x))  
````

## 2. Базовый синтаксис

Ниже — «минимальная грамматика» Python, необходимая для любых дальнейших тем. Важно не просто запомнить правила, а понять, почему они устроены именно так: это поможет быстрее находить ошибки. В отличие от многих языков, Python делает упор на читаемость и использует отступы как часть синтаксиса. Поэтому большая часть типичных проблем у начинающих связана не с математикой, а с форматированием кода и структурой блоков.

### 2.1. Инструкции и выражения
**Выражение** вычисляется в значение, которое можно подставлять в другие выражения: например, `2 + 2`, `x * y`, `f(a)`. 

**Инструкция** выполняет действие и обычно меняет состояние программы: присваивание `x = 3`, условие `if`, цикл `for`, определение функции `def`. На практике вы будете постоянно комбинировать выражения внутри инструкций, например вычислять выражение внутри `print(...)` или внутри `return`.

### 2.2. Отступы
Python использует **отступы** для обозначения блоков кода, то есть для группировки инструкций, которые должны выполняться вместе. Это делает код визуально структурированным, но требует дисциплины: один лишний пробел может изменить выполнение программы или вызвать ошибку. Стандарт де‑факто — одна табуляция на уровень вложенности; Никогда не смешивайте табы и пробелы: внешне код может выглядеть одинаково, но интерпретатор увидит разные отступы.

```
if condition:
    do_something()
    do_more()
```

### 2.3. Комментарии и документация
Комментарии нужны, чтобы пояснять фрагменты кода. Однострочный комментарий начинается с `#` и продолжается до конца строки, его игнорирует интерпретатор. Многострочные строки `"""` часто используют как документацию (docstring) для функций и модулей: такие строки доступны через `help(...)`. Хороший docstring кратко описывает назначение, аргументы и формат результата, что особенно важно в командной работе и при проверке лабораторных.


````python
# Однострочный комментарий: импортируем модуль математики
import math

"""
Многострочный комментарий (часто используют как пояснение блока кода):

Этот пример считает кинетическую энергию по формуле
E = 1/2 * m * v^2

Здесь m — масса (кг), v — скорость (м/с), E — энергия (Дж).
"""

m = 2.0  # масса в килограммах
v = 3.0  # скорость в метрах в секунду

E = 0.5 * m * v**2  # вычисляем энергию
print(f"E = {E} ")  # вывод результата
print("Тип данных E: ", type(E)) # вывод типа данных результата
````

### 2.4. Имена переменных
Имена переменных в Python — это метки (ссылки), которыми вы обращаетесь к объектам, поэтому выбирайте их так, чтобы код читался как текст. Имя может содержать буквы, цифры и `_`, но не может начинаться с цифры; регистр важен (`x` и `X` различны). Зарезервированные слова (`for`, `class`, `def` и др.) использовать нельзя, иначе код не запустится.

### 2.5. Полезные встроенные функции

Встроенные функции помогают быстро «прощупать» объект и понять, как с ним работать, не открывая документацию в браузере. `type(x)` сообщает тип объекта, что особенно важно в динамически типизированном Python. `len(s)` даёт длину строки или контейнера и часто используется в циклах и проверках. `help(obj)` выводит справку и примеры использования, а `dir(obj)` показывает доступные атрибуты и методы — это полезно при исследовательском стиле работы в ноутбуке. Наконец, `print(...)` — базовый инструмент отладки: лучше явно печатать промежуточные значения, чем гадать.


````python
# Типы и вывод
value = 3.14
print(value, "тип:", type(value))
print(help(value))
````

## 3. Типы данных (основы)

Типы данных определяют, какие операции корректны и как интерпретируются значения. В физике типы данных особенно важны, потому что вычисления часто строятся из длинных цепочек формул, и ошибка типа может тихо «сломать» смысл результата. Python даёт удобные базовые типы для чисел, логики и строк, а также контейнеры для группировки данных. На первых шагах полезно постоянно проверять себя функцией `type(...)`, чтобы видеть, что именно получилось после операции.

### 3.1. Числа: `int` и `float`
`int` — это целые числа произвольной точности, то есть Python может хранить очень большие целые без переполнения (но с ростом затрат по времени/памяти). `float` — числа с плавающей точкой (обычно IEEE 754 double), они подходят для большинства инженерных расчётов, но имеют ограниченную точность. Из‑за двоичного представления некоторых десятичных дробей возникают небольшие ошибки, поэтому сравнение `float` «на равенство» часто некорректно. В вычислительной физике это приводит к необходимости использовать допуски (например, `abs(a-b) < eps`) и внимательно относиться к накоплению ошибок округления. Стандартный метод, используемый в вычислительной физике для сравнения чисел с плавающей точкой - функция isclose из модуля math.


````python
"""Вывод представленной математической операции будет False, из за ошибки чисел с плавающей точкой"""
print(0.1 + 0.2 == 0.3)


"""Вывод представленной математической операции будет True используя функцию isclose модуля math"""
import math
print(math.isclose(0.1 + 0.2, 0.3)) 

"""Функция isclose фактически выполняет следующую операцию - сравнение абсолютной разности значений с каким то бесконечно малым epsilon, в данном случе 10^(-12) степени"""
print(abs((0.1 + 0.2) - 0.3) < 1e-12)
````

### 3.2. Логический тип: `bool`

Логический тип имеет два значения: `True` и `False`, и чаще всего получается из сравнений (`x > 0`, `a == b`) или логических выражений. Он используется в условиях `if`, фильтрации данных и критериях остановки итерационных алгоритмов. Операторы `and` и `or` выполняют функции логического "И" и логического "ИЛИ" соответственно. Оператор `not` инвертирует логическое значение и помогает писать ясные отрицания условий. Отличайте `==` (сравнение значений) и `=` (присваивание объектов): это концептуально важно.


````python
a = 5
b = 10
(a < b) and (a > 0), not (a == b)
````

### 3.3. Строки: `str`

Строки — это последовательности символов, которые часто используются для имён файлов, меток, сообщений и обработки текстовых данных эксперимента. Важное свойство: `str` **неизменяем** (immutable), то есть операции «изменения» создают новый объект. Индексация и срезы позволяют извлекать подстроки без сложных циклов, что удобно для парсинга. Форматирование через f‑строки помогает собирать читаемые отчёты и подписи к графикам прямо в коде. При работе с данными полезно помнить о методах строк (`.strip()`, `.split()`, `.replace()`), но сейчас ограничимся базовыми операциями.


````python
s = "physics"
print(s)
print(s[0], s[-1]) #Правила индексации элементов строк
print(s[1:4]) #Правила срезов строк
print(f"Длина слова {s} = {len(s)} символов")
````

### 3.4. `None`

`None` — специальное значение, означающее «отсутствие данных» или «ничего не возвращено». Оно часто встречается, когда функция выполняет действие (например, печать или запись в файл), но не возвращает вычисленное значение. Это удобно для явного обозначения пустых полей в данных или отсутствующих параметров по умолчанию. Проверять `None` следует через `is None`, потому что это проверка на конкретный singleton‑объект, а не сравнение значений. Понимание `None` важно, чтобы не путать «нулевое значение» (`0`) и «отсутствие значения» (`None`).


````python
#Функция, которая ничего не возвращает
def f_no_return():
    x = 1 + 1

result = f_no_return()
print(result, result is None)

#Функция, которая возвращает 1+1
def f():
    return 1 + 1

result2 = f()
print(result2, result2 is None)
````

## 4. Контейнеры и коллекции

Контейнеры позволяют группировать данные и описывать структуры, которые естественно встречаются в физике: набор измерений, таблица параметров модели, множество уникальных идентификаторов и т. п. Отличайте контейнеры по двум особенностям: (1) упорядоченность и доступ по индексу, (2) изменяемость. Эта тема напрямую связана с моделью памяти: изменяемые контейнеры ведут себя иначе при присваивании и передаче в функции. В научном коде контейнеры используются постоянно, даже если позже их заменяют массивами `numpy` для скорости.

### 4.1. Списки: `list` (изменяемые - Mutable), квадратные скобки
Список — это изменяемая упорядоченная коллекция, где элементы могут иметь разные типы (хотя в численных задачах обычно типы однородны). Индексация начинается с нуля, а отрицательные индексы позволяют обращаться к элементам с конца. Методы `append` и `extend` помогают наращивать данные по мере чтения файла или хода эксперимента. Так как список изменяем, любые ссылки на него «видят» изменения, что важно помнить при копировании и передаче в функции. Списки — хороший стартовый инструмент, но для больших массивов чисел в физике обычно переходят к `numpy.ndarray`.


````python
a = [1, 2, 3]
print(a)

a.append(4)
print(a)

a[0] = 10
print(a)
````

### 4.2. Кортежи: `tuple` (неизменяемые - Immutable), круглые скобки

Кортеж похож на список тем, что хранит упорядоченную последовательность, но он **неизменяем**. Это полезно, когда вы хотите зафиксировать «запись» с постоянной структурой: например, координаты точки, параметры эксперимента или пару (время, значение). Неизменяемость повышает надёжность: вы случайно не испортите данные, изменив элемент кортежа. Кортежи также могут использоваться как ключи словаря (если содержат только хэшируемые элементы), что бывает удобно для индексирования по нескольким параметрам. В целом, выбирайте `tuple`, когда структура фиксирована, и `list`, когда вы планируете модификации.


````python
p = (1.0, 2.0, 3.0)
print(p, type(p))

p[0] = 10 # Ошибка, неизменяемый тип данных
````

### 4.3. Словари: `dict`, фигурные скобки, пары "ключ": значение

Словарь хранит пары «ключ → значение» и реализует быстрый доступ по ключу, что делает его естественным выбором для параметров модели или метаданных эксперимента. Ключами обычно служат строки, но могут быть и числа или кортежи (если они хэшируемы). Метод `.get(key, default)` полезен, когда значение может отсутствовать: так вы задаёте безопасное значение по умолчанию и избегаете ошибки `KeyError`. Словари изменяемы, поэтому при передаче в функцию вы можете либо мутировать их «на месте», либо создавать копию, если хотите сохранить исходное состояние.


````python
params = {"key1": 1.0, "key2": 2.0, "key3": 123}
print(params["key1"],"", params["key2"])
print(params.keys())

"""Перебор ключей вместе со значениями в цикле"""
for key, value in params.items():
    print(key, value)
````

### 4.4. Множества: `set`

Множество — это неупорядоченная коллекция **уникальных** элементов, и это его основная ценность. Оно идеально подходит для удаления дубликатов, проверки принадлежности (`x in S`) и операций теории множеств (объединение, пересечение, разность). Проверка принадлежности во множестве обычно быстрее, чем в списке, поэтому `set` часто используют как «таблицу допустимых значений». Элементы множества должны быть хэшируемыми, то есть изменяемые контейнеры вроде списков туда добавить нельзя. В анализе данных множества полезны для сравнения наборов идентификаторов, датчиков, файлов или экспериментальных серий.


````python
A = {1, 2, 3, 3, 2}
B = {3, 4}
print(A, A | B, A & B, A - B)
````

## 5. Управляющие конструкции

Управляющие конструкции задают «динамику» программы: ветвления, повторения и абстракции через функции. В вычислительной физике именно они позволяют реализовать итерационные методы, симуляции по времени и обработку данных. На первых шагах важно писать конструкции максимально прозрачно, даже если код получается длиннее: читаемость сейчас важнее микро‑оптимизаций. Обращайте внимание на границы циклов, условия остановки и то, какие переменные изменяются внутри блока. Эти навыки напрямую переносятся на более сложные темы (например, численные методы решения ОДУ).

### 5.1. Условия `if`
`if` выбирает один из вариантов выполнения в зависимости от логического условия. Блоки `elif` позволяют добавить альтернативные ветви, а `else` служит «последним вариантом», если ни одно условие не сработало. Хорошее условие должно быть простым и проверяемым; если условие становится длинным, его стоит вынести в отдельную переменную с говорящим именем. Важно помнить, что любые выражения в Python имеют «истинность» (truthiness), но на этом этапе лучше явно писать сравнения, чтобы избежать неожиданностей.


````python
x = -2
if x > 0:
    msg = "positive"
elif x == 0:
    msg = "zero"
else:
    msg = "negative"
msg
````

### 5.2. Цикл `for`

Цикл `for` в Python устроен вокруг идеи перебора элементов **итерируемого** объекта, а не вокруг индексов как в некоторых других языках. Это позволяет писать код ближе к математическому описанию: «для каждого элемента из набора сделай ...». `range(n)` создаёт последовательность индексов от `0` до `n-1`, что удобно для задач с фиксированным числом шагов, например дискретизация времени. `enumerate(...)` возвращает пары (индекс, элемент) и помогает избежать ручного управления счётчиком. В научном коде `for` часто используется для накопления суммы, построения таблицы значений или пошагового применения преобразований.


````python
# сумма квадратов 0^2 + 1^2 + ... + (n-1)^2
n = 6
s = 0
for i in range(n):
    s += i**2
s
````

### 5.3. Цикл `while`

`while` повторяет блок, пока условие истинно, и поэтому особенно подходит для методов, где число итераций заранее неизвестно. Примеры из физики: итерационное решение уравнений, поиск корня, сходимость по заданному допуску или имитация процесса до достижения порога. Ключевая опасность — бесконечный цикл, поэтому всегда продумывайте критерий остановки и (часто) добавляйте ограничение на максимальное число шагов. Полезная техника — печатать/логировать промежуточные значения на первых тестах, чтобы убедиться, что алгоритм движется в нужную сторону. Если цикл зависит от `float`, учитывайте численные погрешности и проверяйте остановку с допуском.

### 5.4. Функции

Функция группирует повторяющийся код и превращает его в «чёрный ящик» с понятным интерфейсом: входные аргументы → результат. Это повышает читаемость, облегчает тестирование и позволяет использовать одну и ту же логику во многих местах программы. В физике функции часто отражают формулы (энергия, потенциал, плотность вероятности) или один шаг алгоритма (например, вычисление градиента). Docstring внутри функции служит мини‑документацией и особенно полезен в ноутбуках: её сразу видно через `help(...)`. Привыкайте возвращать значения через `return`, а не печатать их внутри функции, чтобы функцию можно было использовать в вычислениях.


````python
def kinetic_energy(m, v):
    """Кинетическая энергия: E = 1/2 m v^2"""
    return 0.5 * m * v**2

kinetic_energy(2.0, 3.0)
````

## 6. Модель памяти Python: объекты и ссылки

Понимание модели памяти — одно из главных отличий «просто писать код» от «понимать, что происходит». В Python практически всё является объектами, а имена переменных — это ссылки на эти объекты. Из‑за этого присваивание и передача аргументов в функции не работают как «копирование значения» в низкоуровневых языках. Непонимание ссылок приводит к типичным ошибкам: «почему изменился список в другом месте?» или «почему мои данные перезаписались?». В этой секции мы научимся диагностировать такие случаи с помощью `id(...)`, `==` и `is`.

### 6.1. Присваивание не копирует объект
Оператор `=` связывает имя с объектом и сам по себе **не создаёт копию**. Поэтому `b = a` означает, что оба имени указывают на один и тот же объект, и мутация через `b` будет видна через `a`. Чтобы отличать «тот же объект» от «такое же значение», используйте `is` (тождественность) и `==` (равенство значений). Функция `id(obj)` даёт идентификатор объекта (в CPython он обычно связан с адресом в памяти), что полезно для экспериментов и отладки. Важно: полагаться на конкретные значения `id` не нужно, важны только сравнения и факт совпадения/различия.


````python
a = [1, 2, 3]
b = a          # b ссылается на тот же список
b.append(4)

print("a:", a)
print("b:", b)
print("a == b:", a == b)
print("a is b:", a is b)
print("id(a), id(b):", id(a), id(b))
````

### 6.2. Изменяемые и неизменяемые типы

Разделение на mutable и immutable объясняет множество «странных» эффектов в Python. **Неизменяемые** объекты (например `str`, `tuple`) нельзя изменить «на месте», поэтому любая операция, которая выглядит как изменение, на самом деле создаёт новый объект. **Изменяемые** объекты (например, `list`, `dict`, `set`) допускают мутацию: добавление элемента меняет сам объект, и это видят все ссылки на него. В научных расчётах это важно, потому что можно случайно испортить исходные данные, если передать контейнер в функцию и изменить его внутри. Хорошая практика — явно документировать, мутирует ли функция входные данные, и при необходимости делать копии.


````python
x = 10
print("x:", x, "id:", id(x))
x = x + 1
print("x:", x, "id:", id(x))

s = "ab"
print("s:", s, "id:", id(s))
s = s + "c"
print("s:", s, "id:", id(s))
````

### 6.3. Копирование контейнеров: поверхностное и глубокое

Когда вы делаете копию контейнера, нужно понимать, копируются ли вложенные объекты. **Поверхностная копия** (например, `a.copy()` или `a[:]`) создаёт новый контейнер верхнего уровня, но ссылки на вложенные объекты остаются общими. Это означает, что изменения вложенных списков/словарей будут видны и в исходном контейнере, и в копии. **Глубокая копия** (`copy.deepcopy`) рекурсивно копирует вложенные объекты, делая структуру независимой, но это дороже по времени и памяти. В реальном коде выбор между ними зависит от того, собираетесь ли вы мутировать вложенные структуры или нет; лучше выбирать минимально достаточный уровень копирования.


````python
import copy

a = [[1, 2], [3, 4]]
b = a.copy()              # поверхностная копия
c = copy.deepcopy(a)      # глубокая копия

a[0].append(99)           # меняем вложенный список

print("a:", a)
print("b:", b, "  (поверхностная: вложенный объект общий)")
print("c:", c, "  (глубокая: вложенный объект независим)")
print("a[0] is b[0]:", a[0] is b[0])
print("a[0] is c[0]:", a[0] is c[0])
````

### 6.4. Аргументы функций и «передача по ссылке»

В Python корректнее говорить не «передача по ссылке», а *передача по значению ссылки* (call-by-sharing). Функция получает ссылку на объект, и если объект изменяемый, то мутация внутри функции изменит этот объект и будет видна снаружи. Однако если внутри функции вы переприсваиваете имя аргумента на новый объект, это затрагивает только локальную переменную функции и не меняет внешнюю ссылку. Это различие критично при написании численных алгоритмов: иногда вам нужен «in‑place» апдейт для эффективности, а иногда — неизменяемость входных данных для корректности эксперимента. Всегда решайте явно, какой контракт у вашей функции, и при необходимости возвращайте новый объект вместо мутации.


````python
def mutate_list(lst):
    lst.append("X")  # мутация объекта

def rebind_list(lst):
    lst = ["new"]    # перепривязка локального имени
    lst.append("Y")
    return lst

a = [1, 2]
mutate_list(a)
print("после mutate_list:", a)

b = [1, 2]
out = rebind_list(b)
print("после rebind_list b:", b)
print("возвращённый out:", out)
````
